```python
import pandas as pd
score_df = pd.DataFrame([[1,50,80,70,'boy',1], 
              [2,60,45,50,'boy',2],
              [3,98,43,55,'boy',1],
              [4,70,69,89,'boy',2],
              [5,56,79,60,'girl',1],
              [6,60,68,55,'girl',2],
              [7,45,70,77,'girl',1],
              [8,55,77,76,'girl',2],
              [9,25,57,60,'girl',1],
              [10,88,40,43,'girl',3],
              [11,25,60,45,'boy',3],
              [12,80,60,23,'boy',3],
              [13,20,90,66,'girl',3],
              [14,50,50,50,'girl',3],
              [15,89,67,77,'girl',3]],columns=['student_id','math_score','english_score','chinese_score','sex','class'])
score_df
```

```python
#1.找出全年級各科成績最高分與最低分?
print('數學最高分：', score_df.math_score.max())
print('數學最低分：', score_df.math_score.min())
print('英文最高分：', score_df.english_score.max())
print('英文最高分：', score_df.english_score.min())
print('國文最高分：', score_df.chinese_score.max())
print('國文最高分：', score_df.chinese_score.min())
```

```python
score_df.groupby('class').agg(['max', 'min'])
```

```python
print('最高成績：', score_df.max()[1:4])
print('最低成績：', score_df.min()[1:4])
```

```python
score_df['math_score'].max()
score_df.max()['math_score']
score_df[['math_score', 'chinese_score', 'english_score']].max()
```

```python
#2.找出數學班平均最高的班級?
class_mean = score_df.groupby('class')['math_score'].mean().idxmax()
print('數學班平均最高的班級: Class',class_mean)
#print('數學班平均最高的班級: Class', class_mean['math_score'].idxmax())
```

```python
#3.分析全校女生與男生國文平均差幾分?
ch_diff = score_df[['chinese_score', 'sex']].groupby('sex').mean() #要group_by的前面也要挑出來
print(ch_diff)
abs(ch_diff.iloc[0]-ch_diff.iloc[1])
```
